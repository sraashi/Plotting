```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

x = []
y = []
z = []

with open (r".\Documents\3nir.pdb",'r') as file: 
    for rows in file:
        if (rows[0:4] == "ATOM"):
            row = rows.split()
            xx = float(row[6])
            yy = float(row[7])
            zz = float(row[8])
            x.append(xx)
            y.append(yy)
            z.append(zz)


fig = plt.figure()
ax = Axes3D(fig)
ax.plot(x,y,z)

plt.show()

        
    
```
