```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv(r"C:\Users\raash\Documents\HIVout1.csv")

df['chr']= ''
for i in df.index:
    if df['chromosome'][i] == 'X' or df['chromosome'][i] == 'Y':
        df['chr'][i]=0
    else:
        x=int(df['chromosome'][i])
        df['chr'][i]=x

df.head()
plt.scatter(x=df['chr'], y=df['length'])
plt.show()
plt.set_ylabel("Dadad")
plt.set_xlabel("time")

```
